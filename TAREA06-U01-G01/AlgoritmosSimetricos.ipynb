```python
import time
from Crypto.Cipher import AES
from Crypto.Util.Padding import pad, unpad
```

```python
start_file = time.time()

f = open('250w.txt', 'r')
mensajeOriginal = f.read().encode()

end_file = time.time()
print("Tiempo de lectura para el archivo (250 palabras): " + format(end_file-start_file))
print("\r")

start_encrypt = time.time() 

password = b'1234567812345678'
aes = AES.new(password, AES.MODE_ECB)
encrypted_text = aes.encrypt(pad(mensajeOriginal, 32))

end_encrypt = time.time()
print("Tiempo de encriptacion para el archivo (250 palabras): " + format(end_encrypt-start_encrypt))
print(encrypted_text)
print("\r")

start_uncrypt = time.time()

uncrypted = aes.decrypt(encrypted_text)
uncrypted_text = unpad(uncrypted,32)

end_uncrypt = time.time()
print("Tiempo de des-encriptacion para el archivo (250 palabras): " + format(end_uncrypt-start_uncrypt))
print(uncrypted_text)
print("\r")
```

**Algoritmo DES**

```python
import binascii
from Crypto.Cipher import DES

start_file = time.time()

f = open('250w.txt', 'r')
mensajeOriginal = f.read().encode()

end_file = time.time()
print("Tiempo de lectura para el archivo (250 palabras): " + format(end_file-start_file))
print("\r")


start_encrypt = time.time() 

key = b'12345678'	#   (solo 8)
iv = b'12345678'	# Vector de inicialización

# Creó un objeto de cifrado DE DES
cipher1 = DES.new(key, DES.MODE_CFB, iv)

#    
msg = cipher1.encrypt(mensajeOriginal)

end_encrypt = time.time()
print("Tiempo de encriptacion para el archivo (250 palabras): " + format(end_encrypt-start_encrypt))
print(msg)
print("\r")



start_uncrypt = time.time()


# Creó un objeto de descifrado (descifrado cifrado no puede usar la misma clave)
cipher2 = DES.new(key, DES.MODE_CFB, iv)

# Proceso de descifrado
uncrypted_text = cipher2.decrypt(msg).decode()

end_uncrypt = time.time()
print("Tiempo de des-encriptacion para el archivo (250 palabras): " + format(end_uncrypt-start_uncrypt))
print(uncrypted_text)
print("\r")
```
